```python
import asyncio

from browser_use import Browser as BrowserUseBrowser
from browser_use import BrowserConfig
from browser_use.browser.context import BrowserContext, BrowserContextConfig
```

```python
from googlesearch import search

query = "2025 f1 australia result"
loop = asyncio.get_event_loop()
links = await loop.run_in_executor(
    None, lambda: list(search(query, num_results=10))
)
```

```python
links
```

```python
browser = BrowserUseBrowser(BrowserConfig(headless=False))
```

```python
browser_context = await browser.new_context()
```

```python
await browser_context.navigate_to(links[8])
```

```python
await browser_context.get_page_html()
```

```python
await browser_context.execute_javascript("document.body.innerText")
```

```python
page  = await browser_context.get_current_page()
html_content = await page.content()
```

```python
import markdownify
content = markdownify.markdownify(html_content)
```

```python
content
```

```python
session = await browser_context.get_session()
session.cached_state = await browser_context._update_state()
```

```python
session.cached_state.selector_map
```

```python
await browser.close()
```
